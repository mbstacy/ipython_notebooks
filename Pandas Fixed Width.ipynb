```python
import pandas as pd
wdths = [2,9,8,9,7,9,8,8,7,6]
names = ['day','pool-elevation-0800-ft','pool_elevation-2400-ft','storage-ac-ft','release-power-dsf','release-total-dsf',
         'evap-in','inflow-adj-dsf','rainfall-dam','rainfall-bsn']
def get_data(url):
    df = pd.read_fwf(url,names=names,widths=wdths,skiprows=17,skipfooter=15)
    return  df[pd.notnull(df['pool-elevation-0800-ft'])]
```

```python
from datetime import datetime
months =['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Nov','Dec']
start = 1990
stop = 2014
site= 'SKIA'
template = 'http://www.swt-wc.usace.army.mil/chartbin/%s%s%s.shtml'
urls=[]
pieces =[]
for yr in range(start,stop):
    for m in months:
        try:
            url = template % (site,m,str(yr)[2:4])
            df = get_data(url)
            d = datetime(yr,months.index(m)+1,df['day'])
            #print d
            df['monthyear'] = d
            pieces.append(df)
        except:
            pass
        
#df = get_data('http://www.swt-wc.usace.army.mil/chartbin/SKIAJul97.shtml')
df['pool_elevation-2400-ft'].mean()
#df
```
