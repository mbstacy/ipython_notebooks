```python
from netCDF4 import Dataset

var_name = 'tas'
nc = Dataset('/Users/mstacy/HadGEM2-ES.r1i1p1.tas.rcp26.day.2006.2100.nc')
var = nc.variables[var_name]
var.shape
```

```python
%matplotlib inline
def decdeg2dms(dd):
   is_positive = dd >= 0
   dd = abs(dd)
   minutes,seconds = divmod(dd*3600,60)
   degrees,minutes = divmod(minutes,60)
   degrees = degrees if is_positive else -degrees
   return (degrees,minutes,seconds)
```

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import netCDF4

lat = -97.358
lon = 35.195

lat = decdeg2dms(lat)[0]
lon = decdeg2dms(lon)[0]

print lat, lon
                 
latvar = nc.variables['lat']
lonvar = nc.variables['lon']
print latvar.units, lonvar.units
h = var
times = nc.variables['time']
jd = netCDF4.num2date(times[:],times.units)


hs = pd.Series(h[:,(np.abs(latvar[:]-(lat))).argmin(),(np.abs(lonvar[:]-(lon))).argmin()],index=jd)
```

```python
fig = plt.figure(figsize=(12,4))
ax = fig.add_subplot(111)
ax.set_ylabel('Degrees ' + h.units)
loc = str(latvar[(np.abs(latvar[:]-(lat))).argmin()]) + ',' + str(lonvar[(np.abs(lonvar[:]-(lon))).argmin()])
hs.plot(ax=ax,title='%s at Location %s (Closest Grid Cell to %s,%s)' % (h._name,loc,str(lat),str(lon)))
```
