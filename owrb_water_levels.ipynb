```python
import pandas as pd
list_df =pd.read_html('http://www.owrb.ok.gov/wd/search_test/water_levels.php?siteid=94941',skiprows=1)
df = list_df[0]
df.columns =['date','measurement','status','project']
#df['year'] = [t.year for t in df.date]
#df['month'] = [t.month for t in df.date]
grouped =df.groupby(df.date.map(lambda x: "%d_%02d" % (x.year,x.month))) # ['year','month'])
grouped['measurement'].agg(['min', 'mean', 'max','size'])
```

```python
%matplotlib inline
grouped.plot()
```
