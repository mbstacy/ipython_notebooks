### OWS: Water usage by Oil Gas

```python
from pandas import read_excel, concat, merge, options,unique
#Display options set for pandas 
options.display.max_columns = 50
options.display.max_rows = 250
H_list=[]
tt_list=[]
template= '/Users/mstacy/Desktop/ows-data/Export - Excel Well Workbook_%d.xlsx'
for num in [1,2,3]:
    Header = read_excel(template % (num),sheetname='Header')
    H_list.append(Header)
    test_treat = read_excel(template % (num),sheetname='Test Treatment')
    tt_list.append(test_treat)

Header=concat(H_list)
test_treat= concat(tt_list)

#loaded from all excel speadsheets the Header Tab and the Test Treatment Tabs
```

```python
#merge and display want columns
df = merge(Header,test_treat,on='UWI')
result=df[['UWI','Date Completion','Fluid Amount (gals)','Type','Fluid Amount','Fluid Units','Fluid Type']]
#result DataFrame
#show unique Fluid Types
unique(result['Fluid Type'])
```

```python
from pandas import notnull
#Exclude Fluid Types that are not Water(~ is the NOT operator)
result_a = result[~result['Fluid Type'].isin(['NTGN','A','ACID','O','FFR','F','DOIL','OIL','CO2','G','MA','NAF','HEAT','KERO','BDA'])]
#remove NA values from Fluid Amount
result_a = result_a[notnull(result_a['Fluid Amount (gals)'])]
grouped = result_a.groupby([result_a['Date Completion'].map(lambda x: x.year)])
gby_DC_results=grouped['Fluid Amount (gals)'].agg(['sum','min','mean','max','size'])
#export to csv
#gby_DC_results.to_csv('/Users/mstacy/Desktop/ows-data/result_water_usage_groupbyYear.csv')

#display gby_DC_results
gby_DC_results
```

```python
#Group by Date Completion and Fluid Type
grouped = result_a.groupby([result_a['Date Completion'].map(lambda x: x.year),result_a['Fluid Type']])
gby_DC_FT_results=grouped['Fluid Amount (gals)'].agg(['sum','min','mean','max','size'])
#Export to CSV
#gby_DC_FT_results.to_csv('/Users/mstacy/Desktop/ows-data/result_water_usage_groupbyYearFluidType.csv')

#Display gby_DC-FT_results
gby_DC_FT_results
```
