```python
import netCDF4
filename = '/Users/mstacy/MIROC-ESM.r1i1p1.tas.hist.day.1900.2005.nc'
nc = netCDF4.Dataset(filename, 'r')
#print ncfile                # shows global attributes, dimensions, and variables
ncvars = nc.variables   # a dictionary of variables
print ncvars
# print information about specific variables, including type, shape, and attributes
#for varname in ['temperature', 'salinity', 'Latitude', 'Longitude']:
#    print ncvars[varname]
```

```python
print ncvars['lon']
print ncvars['lat']
var_name = 'tas'
var = nc.variables[var_name]
var.shape
```

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
#import netCDF4
lat = -160.358
lon = 65.195

latvar = nc.variables['lat']
lonvar = nc.variables['lon']
print latvar.units, lonvar.units
h = var
times = nc.variables['time']
jd = netCDF4.num2date(times[:],times.units)


hs = pd.Series(h[:,(np.abs(latvar[:]-(lat))).argmin(),(np.abs(lonvar[:]-(lon))).argmin()],index=jd)
```

```python
%matplotlib inline
fig = plt.figure(figsize=(12,4))
ax = fig.add_subplot(111)
ax.set_ylabel('Degrees ' + h.units)
loc = str(latvar[(np.abs(latvar[:]-(lat))).argmin()]) + ',' + str(lonvar[(np.abs(lonvar[:]-(lon))).argmin()])
hs.plot(ax=ax,title='%s at Location %s (Closest Grid Cell to %s,%s)' % (h._name,loc,str(lat),str(lon)))
```
