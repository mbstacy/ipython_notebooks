```python
import json, pandas as pd
#from pandas import read_csv, concat

url= 'http://waterservices.usgs.gov/nwis/site/?format=rdb&stateCd=ok&siteStatus=%s'
names=['agency_cd','site_no','station_nm','site_tp_cd','dec_lat_va','dec_long_va',
       'coord_acy_cd','dec_coord_datum_cd','alt_va','alt_acy_va','alt_datum_cd','huc_cd']
pieces=[]
for state in ['active','inactive']:
    df=pd.read_csv(url % (state),names=names,skiprows=32,sep="\t")
    df['status']=state.title()
    pieces.append(df)
df = pd.concat(pieces)
```
