```python
from pandas import read_excel, concat, merge, options
#Display options set for pandas 
options.display.max_columns = 50
options.display.max_rows = 250
H_list=[]
tt_list=[]
template= '/Users/mstacy/Desktop/ows-data/Export - Excel Well Workbook_%d.xlsx'
for num in [1,2,3]:
    Header = read_excel(template % (num),sheetname='Header')
    H_list.append(Header)
    test_treat = read_excel(template % (num),sheetname='Test Treatment')
    tt_list.append(test_treat)

Header=concat(H_list)
test_treat= concat(tt_list)
```

```python
#merge and display want columns
df = merge(Header,test_treat,on='UWI')
result=df[['UWI','Date Completion','Fluid Amount (gals)','Type','Fluid Amount','Fluid Units','Fluid Type']]
from pandas import unique
unique(result['Fluid Type'])
```

```python
import pandas as pd

result_a = result[~result['Fluid Type'].isin(['NTGN','A','ACID','O','FFR','F','DOIL','OIL','CO2','G','MA','NAF','HEAT','KERO','BDA'])]
result_a = result_a[pd.notnull(result_a['Fluid Amount (gals)'])]
grouped = result_a.groupby([result_a['Date Completion'].map(lambda x: x.year)])
g4=grouped['Fluid Amount (gals)'].agg(['sum','min','mean','max','size'])
g4.to_csv('/Users/mstacy/Desktop/ows-data/result_water_usage_groupbyYear.csv')
g4
```

```python
grouped = result_a.groupby([result_a['Date Completion'].map(lambda x: x.year),result_a['Fluid Type']])
g3=grouped['Fluid Amount (gals)'].agg(['sum','min','mean','max','size'])
g3.to_csv('/Users/mstacy/Desktop/ows-data/result_water_usage_groupbyYearFluidType.csv')
g3
```
