```python
from urllib2 import urlopen
from pandas import read_fwf, read_csv, concat
import matplotlib.pyplot as plt
import numpy as np

url ="http://www.waterqualitydata.us/Result/search?siteid=ABSHAWNEE-LR02&mimeType=csv"
page = urlopen(url)
df= read_csv(page)
grouped = df.groupby('CharacteristicName')
grouped['ResultMeasureValue'].agg([np.min, np.mean, np.max,np.size])
```

```python
%matplotlib inline
for key, grp in grouped:
    plt.plot(grp['ResultMeasureValue'], label=key)
    
plt.legend(loc='best')    
plt.show()
```
