```python
from pandas import read_excel, concat, merge, options,unique, melt
#from pandas.io.parsers import ExcelFile
#Display options set for pandas 
options.display.max_columns = 50
options.display.max_rows = 50

template= 'GroundwaterDB_Parsons.xlsm'
#Load Spreadsheet
test_treat = read_excel(template,sheetname='Analyses Entry Master Sheet',header=2)
column_names = test_treat.columns

#Need to add a check to see if columns names changed!!!
```

```python
import numpy as np
#Rename Column Names
names = [u'Record Number', u'SampleNum', u'EPA Activity ID', u'Unnamed: 3', u'Waterbody description', u'Waterbody Type', u'EPA Well Type', u'EPA MonLocType', u'Unnamed: 8', u'Depth_ft', u'Sample depth SHALLOWEST interval value', u'Sample depth DEEPEST interval value', u'Waterbody depth (number only)', u'Shallow; Deep≥25', u'Unnamed: 14', u'Location Description', u'Surface Owner', u'Operator', u'Well', u'Location Comments', u'Name description', u'Unique Station ID', u'EPA Proj Sampling Design Types', u'Category', u'GPS How', u'Lat', u'Long', u'Quarters', u'LegalLoc', u'County', u'Dist', u'Nearest Stream', u'Unnamed: 32', u'Sample Date', u'Sample Time', u'What test', u'TDS, mg/l\n(500 scndry)', u'Unnamed: 37', u'Na ppm\n(250 mg/l) ', u'Ca ppm ()', u'Mg ppm ()', u'K ppm ()', u'Nitrate-N ppm\n(10 mg/l)', u'Cl ppm\n(250 mg/l)', u'SO4 ppm\n(250 mg/l)', u'Boron ppm\n(1 ppm OCC)', u'HCO3 ppm ()', u'pH\nstdunit (<6.5 or >8.5)', u'Cond uMoh/cm ()', u'TotlSolSlts (500 scndry)', u'Unnamed: 50', u'Br ppm ()', u'Oil Seen (Y/N NR = not recorded)', u'Toluene ug/l\n(1 mg/l, 1000 ug/l)', u'Benzene ug/l\n(5 ug/l)', u'Ethylb ug/l\n(.7 mg/l, 700 ug/l)', u'Xylenes ug/l\n(10 mg/l, 10,000 ug/l)', u'Naphthalene ug/L ()', u'GRO ug/l\n(25 mg/l)', u'DRO mg/l\n(25 mg/l)', u'6-12 =GRO\n(25 mg/l)', u'12-28 =DRO\n(25 mg/l)', u'28-36\n(25 mg/l)', u'6-36\n(25 mg/l)', u'Unnamed: 64', u'As mg/l\n(.01 mg/l)', u'Ba mg/l\n(2 mg/l)', u'Cr mg/l\n(.1 mg/l)', u'Pb mg/l\n(.015 mg/l)', u'Unnamed: 69', u'Na/Cl\noilfild_(<.6)', u'Br/Cl\noilfield_(>.0035)', u'Unnamed: 72', u'Sample Comments', u'In ArcMap']
temp=[]
for name in names:
    temp.append(name.strip().replace(' ','_').replace('\n','_'))
test_treat.columns=temp

#Remove Unnamed Columns
cols = [c for c in test_treat.columns if c[:7] != 'Unnamed']
df=test_treat[cols]

#Set Location ID
df['Location_id'] = df.Lat.map(str)  + df.Long.values.astype(str)
df['Location_id']=df['Location_id'].apply(lambda x : x.replace('-','_'))
#Sample Dataframe
df_sample=df[['EPA_Activity_ID','Location_id','Sample_Date', u'Sample_Time', u'What_test',u'Depth_ft', u'Sample_depth_SHALLOWEST_interval_value', u'Sample_depth_DEEPEST_interval_value']]
df_sample.columns=['EPA_Activity_ID','Location_id','Sample_Date', u'Sample_Time', u'Test_Type',u'Depth_ft', u'Sample_depth_SHALLOWEST_interval_value', u'Sample_depth_DEEPEST_interval_value']
#Results Dataframe
df_results = melt(df,id_vars=['EPA_Activity_ID'],value_name='value',value_vars=[ u'TDS,_mg/l_(500_scndry)', u'Na_ppm_(250_mg/l)', u'Ca_ppm_()', u'Mg_ppm_()', u'K_ppm_()', u'Nitrate-N_ppm_(10_mg/l)', u'Cl_ppm_(250_mg/l)', u'SO4_ppm_(250_mg/l)', u'Boron_ppm_(1_ppm_OCC)', u'HCO3_ppm_()', u'pH_stdunit_(<6.5_or_>8.5)', u'Cond_uMoh/cm_()', u'TotlSolSlts_(500_scndry)', u'Br_ppm_()', u'Oil_Seen_(Y/N_NR_=_not_recorded)', u'Toluene_ug/l_(1_mg/l,_1000_ug/l)', u'Benzene_ug/l_(5_ug/l)', u'Ethylb_ug/l_(.7_mg/l,_700_ug/l)', u'Xylenes_ug/l_(10_mg/l,_10,000_ug/l)', u'Naphthalene_ug/L_()', u'GRO_ug/l_(25_mg/l)', u'DRO_mg/l_(25_mg/l)', u'As_mg/l_(.01_mg/l)', u'Ba_mg/l_(2_mg/l)', u'Cr_mg/l_(.1_mg/l)', u'Pb_mg/l_(.015_mg/l)'])
#Location Dataframe
df_location=df[['Location_id','Name_description',u'GPS_How', u'Lat', u'Long', u'Quarters', u'LegalLoc', u'County', u'Dist', u'Nearest_Stream','Location_Description','Location_Comments','Surface_Owner','Operator','Well','EPA_MonLocType','EPA_Well_Type','Waterbody_Type','Waterbody_description',]]
df_location.columns=['Location_id','Name_description',u'GPS_How', u'Lat', u'Long', u'Quarters', u'LegalLoc', u'County', u'Distance_Stream', u'Nearest_Stream','Location_Description','Location_Comments','Surface_Owner','Operator','Well','EPA_MonLocType','EPA_Well_Type','Waterbody_Type','Waterbody_description',]
df_location.drop_duplicates(cols = 'Location_id', inplace = True)
#Result Dataframe
df_results["id"] = df_results.index
df_results["parameter"]=df_results.variable.apply(lambda x: x.split('_')[0].replace(',',''))
df_results["unit"]=df_results.variable.apply(lambda x: x.split('_')[1])
df_results["regulation"]=df_results.variable.apply(lambda x: x.split('(')[1][:-1].replace('_',' '))
#Remove Null
df_result=df_results[df_results.value.notnull()]
df_result=df_result[["EPA_Activity_ID","id","parameter","value","unit","regulation","variable"]]
df_result.columns = ["EPA_Activity_ID","id","parameter","value","unit","regulation","wide_column_name"]

#Export to_csv
df_location.to_csv('dt_location.csv',encoding='utf-8')
df_sample.to_csv('dt_sample.csv',encoding='utf-8')
df_result.to_csv('dt_result.csv',encoding='utf-8')

df_result
```
